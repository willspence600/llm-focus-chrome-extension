```python
# Save the system prompt as a variable and use that instead of this beefy text here
#TEST OUT RANDOM WEBSITES BC YOUTUBE DOESNT HAVE MUCH TEXT
# use the following format for testing cases and determining accuracy:
    #{"Webpage Content": {"AI": 1, "Cooking": 0}}
    #{"Webpage Content": {"AI": 1, "Cooking": 0}}
#Also add in a part about search engines or general sites and not to block them until the user seems to have reached their desired page/a more specific page
#Add in a part where it errs on the side of caution where if something is close or it can't really determine the relevance, or it has minimal relevance but still relevance, it outputs TRUE
    #Maybe this is a scale ranking system where it ranks the relevance of the website and only outputs TRUE for a certain relevance score or higher

#Get a python script we can run

import cohere, ast, re
from system_prompts.prompt_v1 import how_prompt
# from prompt-v1 in system-prompts/ import how_prompt

co = cohere.ClientV2("85pa1ijYVknf5xRslpLJ5lB0sBCUXmDyyEPSoQKn")
response = co.chat(
    model="command-r-plus", 
    messages=[{"role": "system", "content": how_prompt},
              {"role": "user", "content": "I want to study coding for 3 hours with a 20 minute break every 25 minutes starting at 15 minutes."}]

)
#I want to study coding for 30 minutes.

input_string = response.message.content[0].text

# Clean the string to match proper Python dictionary format:
# - Add quotes around keys
cleaned_string = re.sub(r'(\w+)(\s*:\s*)', r'"\1"\2', input_string.strip())

# Convert the cleaned string into a Python dictionary
session_details = eval(cleaned_string)

# Function to generate the study and break schedule as a list of tuples
def create_activity_schedule(session_details):
    len_session = session_details["len_session"]
    num_breaks = session_details["num_breaks"]
    len_breaks = session_details["len_breaks"]
    first_break = session_details["first_break"]
    break_interval = session_details["break_interval"]
    
    # Create the activity schedule
    activities = []
    current_time = 0  # Start at the beginning of the session
    
    # Then alternate between study and break sessions
    for i in range(num_breaks):  # Subtract 1 because we already added the first break
        if current_time == 0:
            activities.append(("study", first_break))
            current_time += first_break 

        if len_session - current_time == 0:
            return activities
        
         # Add break time after the study period
        if current_time < len_session:  # Avoid adding break if it exceeds session time
            activities.append(("break", min(len_session - current_time, len_breaks)))
            current_time += min(len_session - current_time, len_breaks)
        
        if len_session - current_time == 0:
            return activities
        
        # Add study time
        activities.append(("study", min(len_session - current_time, break_interval)))
        current_time += min(len_session - current_time, break_interval)
        
       
        
    # If there's remaining time after all breaks, add a final study period
    remaining_time = len_session - current_time
    print(remaining_time)
    if remaining_time > 0:
        activities.append(("study", remaining_time))
    
    return activities

# Generate the activity schedule
activity_schedule = create_activity_schedule(session_details)

# Print the result
#for activity in activity_schedule:
#    print(activity)

```
